```python
# y=a+bx


def sum_of_squred_error(xdata,ydata,a,b):
    total_error=0
    for x,ya in zip(xdata,ydata):
        yp=a+b*x
        error=(1/2)*(yp-ya)**2
        total_error=total_error+error
    return total_error

xd=[1,2,3,4]
yd=[10,12,13,14]
a=1
b=2

print(sum_of_squred_error(xd,yd,a,b))
```
