```python
# y=a+bx



def sum_of_squred_error(xdata,ydata,a,b):
    total_error=0
    for (x,ya) in zip(xdata,ydata):
        yp=a+b*x
        error=(1/2)*(yp-ya)**2
        total_error=total_error+error
    return total_error


a=0.45
b=0.75

print(sum_of_squred_error(df['Xa'],df['Ya'],a,b))
```

```python
# y=a+bx

a=0.45
b=0.75

def decrease_gradient_value(xdata,ydata,a,b):
    r=0.01
    total_error=0
    sum_a=0
    sum_b=0
    for (x,ya) in zip(xdata,ydata):
       
        yp=a+b*x
        sum_a= sum_a + (-(ya-yp))
        sum_b= sum_b + (-(ya-yp)*x)

        new_a=a-r*sum_a
        new_b=b-r*sum_b
        error=(1/2)*(yp-ya)**2
        total_error=total_error+error
    return total_error,new_a,new_b




print(sum_of_squred_error(df['Xa'],df['Ya'],a,b))
```
