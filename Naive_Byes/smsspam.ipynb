```python
import string
# Define the function to remove the punctuation
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
# Apply to the DF Columns
df['result'] = df['result'].apply(remove_punctuations)
```

```python
# Term Document Matrix (TDM) for spam Class
```

```python
from sklearn.feature_extraction.text import CountVectorizer

spam_sms = [row['result'] for index,row in df.iterrows() if row['type'] == 'spam']
vec_s = CountVectorizer()
X_s = vec_s.fit_transform(spam_sms)
tdm_spam = pd.DataFrame(X_s.toarray(), columns=vec_s.get_feature_names())

tdm_spam
# print(type(X_s))
```

To find the total number of times a word appears in a class, we can use CountVectorizer from sklearn.
CountVectorizer gives Term-Document Matrix (TDM) for each class. A term-document matrix (TDM) consists of a list of word frequencies appearing in a set of documents.

```python
# Term Document Matrix (TDM) for ham Class
```

```python
ham_sms = [row['result'] for index,row in df.iterrows() if row['type'] == 'ham']

vec_h= CountVectorizer()
X_h = vec_h.fit_transform(ham_sms)
tdm_ham = pd.DataFrame(X_h.toarray(), columns=vec_h.get_feature_names())

tdm_ham
```

```python
#Frequency of words for Spam Class:
```

```python
word_list_s = vec_s.get_feature_names();    
count_list_s = X_s.toarray().sum(axis=0) 

freq_s = dict(zip(word_list_s,count_list_s))
freq_s
```

```python
#Similarly, Frequency of words for ham Class
```

```python
word_list_h = vec_h.get_feature_names();    
count_list_h = X_h.toarray().sum(axis=0) 

freq_h = dict(zip(word_list_h,count_list_h))

freq_h
```

```python
#Probabilities of words in Spam Class:
```

```python
prob_s = []
for word, count in zip(word_list_s,count_list_s):
    prob_s.append(count/len(word_list_s))
dict(zip(word_list_s,prob_s))
```

```python
#Probabilities of words in ham  Class:
```

```python
prob_h = []
for count in count_list_h:
    prob_h.append(count/len(word_list_h))
dict(zip(word_list_h,prob_h))
```

```python
from nltk.tokenize import word_tokenize
new_sms = 'Free entry in  a wkly comp to win FA Cup final tkts st May Text FA to  to receive entry question std txt rate T Cs apply overs'
new_word_list = word_tokenize(new_sms)
print (new_word_list)
```

```python
prob_s_with_ls = []
for word in new_word_list:
    if word in freq_s.keys():
        count = freq_s[word]
    else:
        count = 0
    prob_s_with_ls.append((count + 1)/(total_sms_features_s + total_features))
dict(zip(new_word_list,prob_s_with_ls))
```

```python
spam_prob_dict=dict(zip(new_word_list,prob_s_with_ls))
```

```python
prob_spam=1

for i in spam_prob_dict.values():
    prob_spam=prob_spam * i
prob_spam  
```

```python
prob_h_with_ls = []
for word in new_word_list:
    if word in freq_h.keys():
        count = freq_h[word]
    else:
        count = 0
    prob_h_with_ls.append((count + 1)/(total_sms_features_h + total_features))
dict(zip(new_word_list,prob_h_with_ls))
```

```python
ham_prob_dict=dict(zip(new_word_list,prob_h_with_ls))
```

```python
prob_ham=1

for i in spam_prob_dict.values():
    prob_ham=prob_ham * i

prob_ham  
```

```python
import string
Punct_to_remove=string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('','',Punct_to_remove))

df['result']=df['result'].apply(lambda text:remove_punctuation(text))
df
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')


STOPWORDS = set(stopwords.words('english'))

def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["result"] = df["result"].apply(lambda text: remove_stopwords(text))


df.head()
```

```python
x=df['result']
y=df['type']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
x=cv.fit_transform(df['result']).toarray()
x
```

```python
y = pd.get_dummies(df['type'])
y = y.iloc[:,1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)

from sklearn.naive_bayes import MultinomialNB
spam_detection_model=MultinomialNB().fit(x_train,y_train)

```

```python
y_pre=spam_detection_model.predict(x_test)

```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
import seaborn as sns
```

```python
acc=accuracy_score(y_test,y_pre)
acc
```
